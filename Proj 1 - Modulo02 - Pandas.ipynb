## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50


```python
# Criar coluna "Total" somando o valor de todos os ativos por investidor
lista_col = ['Ativo01' , 'Ativo02' , 'Ativo03' , 'Ativo04' , 'Ativo05' , 'Ativo06' , 'Ativo07']

df_inv_cad['Total'] = df_inv_cad[lista_col].sum(axis=1)
df_inv_cad
```

```python
# Criar coluna "Media" com a média das notas 1 e 2
lista_col_2 = ['Nota1' , 'Nota2']
df_inv_cad['Media'] = df_inv_cad[lista_col_2].mean(axis=1)
df_inv_cad
```

```python
# Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
df_inv_cad[['nome','dominio']] = df_inv_cad['Email'].str.split('@', expand=True)
df_inv_cad
```

```python
# Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
df_inv_cad = df_inv_cad.rename(columns = {'Salario': 'Renda'})
df_inv_cad
```

```python
# criar nova coluna de renda só que SEM o carcter $
df_inv_cad['renda_ajust'] = df_inv_cad['Renda'].str.replace('$','')
df_inv_cad
```

```python
df_inv_cad['renda_ajust']
```

```python
# CONVERTER a nova coluna renda_ajust de OBJECT em FLOAT
pd.to_numeric(df_inv_cad['renda_ajust'])
```

```python
# Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
df_inv_cad['Sexo'] = df_inv_cad['Sexo'].map({'Masc': 'Masculino', 'Fem':'Feminino'})
```

```python
df_inv_cad
```

```python
# Transformar os dados da coluna Media em inteiro
df_inv_cad['Media'].astype(int)
```

```python
# Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, 
# como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50
# df_inv_cad['NPS'] = pd.cut(df_inv_cad['Media'],3,labels=['detrator', 'neutro', 'promotor'])
# (0,49), (50,79), (80,200)

bins = pd.IntervalIndex.from_tuples([(0,49), (50,79), (80,200)]) # criei uma tupla com os intervalos possiveis
df_inv_cad['NPS_intervalo'] = pd.cut(df_inv_cad['Media'], bins) # criar coluna transitória NPS_intervalo com os valores das tuplas
```

```python
 # verificar o tipo do novo campo
df_inv_cad['NPS_intervalo']
```

```python
# converter category ==> inteiro
df_inv_cad['NPS'] = df_inv_cad['NPS_intervalo'].cat.codes
```

```python
 # verificar se alterou o novo campo
df_inv_cad['NPS']
```

```python
# alterar os inteiros (0 ==> detrator, 1 ==> neutro, 2 ==> promotor)
df_inv_cad['NPS'] = df_inv_cad['NPS'].map({0:'detrator', 1:'neutro', 2:'promotor'})
df_inv_cad
```
